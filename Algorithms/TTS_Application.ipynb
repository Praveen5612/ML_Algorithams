```python
import streamlit as st
from gtts import gTTS
import pyttsx3
import os

# Function to generate audio using Google Text-to-Speech (gTTS)
def generate_audio_gtts(text, lang='en'):
    """Generate audio using Google Text-to-Speech."""
    tts = gTTS(text=text, lang=lang)
    file_path = "generated_audio_gtts.mp3"
    tts.save(file_path)
    return file_path

# Function to generate audio using pyttsx3
def generate_audio_pyttsx3(text, rate=150, volume=1.0):
    """Generate audio using pyttsx3."""
    engine = pyttsx3.init()
    engine.setProperty('rate', rate)  # Speed of speech
    engine.setProperty('volume', volume)  # Volume level (0.0 to 1.0)
    file_path = "generated_audio_pyttsx3.mp3"
    engine.save_to_file(text, file_path)
    engine.runAndWait()
    return file_path

# Streamlit app interface
def run_app():
    """Main function to run the Streamlit app."""
    st.title("Text-to-Speech Generator")

    # Get user input
    text = st.text_area("Enter text to convert to speech:")
    language = st.selectbox("Select Language", ["en", "es", "fr", "de", "it"])
    engine_choice = st.radio("Choose Text-to-Speech Engine", ("gtts", "pyttsx3"))
    rate = st.slider("Speech Rate", 50, 250, 150)
    volume = st.slider("Volume", 0.0, 1.0, 1.0)

    # Generate audio on button click
    if st.button("Generate Audio"):
        if not text.strip():
            st.error("Text input is required.")
        else:
            try:
                if engine_choice == "gtts":
                    audio_file = generate_audio_gtts(text, lang=language)
                elif engine_choice == "pyttsx3":
                    audio_file = generate_audio_pyttsx3(text, rate=rate, volume=volume)
                else:
                    st.error("Invalid TTS engine choice.")
                    return

                # Provide download link for generated audio
                st.success("Audio generated successfully!")
                st.audio(audio_file, format="audio/mp3")
                st.download_button("Download Audio", audio_file, file_name="generated_audio.mp3", mime="audio/mp3")

            except Exception as e:
                st.error(f"Error: {str(e)}")

# Run the Streamlit app
if __name__ == "__main__":
    run_app()
```
