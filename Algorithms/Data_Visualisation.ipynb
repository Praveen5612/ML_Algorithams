```python
from scipy import stats
import numpy as np

z_scores = np.abs(stats.zscore(data[['energy']].dropna()))
filtered_data = data[(z_scores < 2).all(axis=1)]

print("Original shape:", data.shape)
print("After removing outliers:", filtered_data.shape)
```
